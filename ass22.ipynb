```python
#Q2.

#In pandas, both loc and iloc are used to access and manipulate data in a DataFrame, but they have different ways of indexing and selecting data.

#    loc:
#        loc is primarily label-based indexing. It is used to select rows and columns based on their labels (row and column names).
#        It uses row and column labels to index data. This means you provide the actual row and column labels to extract data.
#        You can use boolean arrays along with loc to filter data based on conditions.
#        Example: df.loc[2, 'column_name'] selects the value in the row with label 2 and the column with the label 'column_name'.

#   iloc:
#      iloc is primarily integer-based indexing. It is used to select rows and columns based on their integer positions (row and column indices).
#      It uses integer indices to index data. This means you provide the integer positions to extract data.
#      You can use slices and integer lists along with iloc to select specific rows and columns.
#      Example: df.iloc[2, 3] selects the value in the third row and fourth column (0-based indices).

#It's important to note that loc includes both the start and end labels in slicing, while iloc follows the usual Python convention of including the start index but excluding the end index in slicing.

import pandas as pd

data = {'A': [1, 2, 3], 'B': [4, 5, 6], 'C': [7, 8, 9]}
df = pd.DataFrame(data, index=['row1', 'row2', 'row3'])

# Using loc
print(df.loc['row2', 'B'])         
print(df.loc['row1':'row2', 'A'])  

# Using iloc
print(df.iloc[1, 1])              
print(df.iloc[0:2, 0])           
```

```python
#Q3.


import pandas as pd

# Sample dataframe
data = {'A': [10, 20, 30, 40], 'B': [1, 2, 3, 4]}
df = pd.DataFrame(data)

# Reindexing
reindex = [3, 0, 1, 2]
new_df = df.reindex(reindex)

# Outputs
output_loc = new_df.loc[2]
output_iloc = new_df.iloc[2]

print("Output for new_df.loc[2]:")
print(output_loc)

print("\nOutput for new_df.iloc[2]:")
print(output_iloc)

#Now, let's discuss the difference in the outputs for new_df.loc[2] and new_df.iloc[2]:

#    new_df.loc[2]:
#    This will return the row with the label/index 2 from the new_df dataframe. Since you've reindexed the dataframe using the reindex list, the index labels have changed. In the original df, the row with index 2 had the values [30, 3]. However, after reindexing, the row with label 2 will correspond to the row that was originally at index 1 in the df, which had the values [20, 2].

#    new_df.iloc[2]:
#    This will return the third row from the new_df dataframe using integer-based indexing (i.e., positional indexing). In this case, the third row in new_df corresponds to the row that was originally at index 1 in the original df, which had the values [20, 2].

#So, while the values in new_df.loc[2] and new_df.iloc[2] will be the same (both will be [20, 2]), the difference lies in how they are retrieved. loc uses label-based indexing, whereas iloc uses integer-based positional indexing. This demonstrates the impact of reindexing on the behavior of loc and iloc when fetching rows from the dataframe.
```

```python
#Q8.

import pandas as pd

def calculate_time_difference(date1, date2):
    # Convert input strings to datetime objects
    datetime1 = pd.to_datetime(date1)
    datetime2 = pd.to_datetime(date2)
    
    # Calculate time difference
    time_difference = datetime2 - datetime1
    
    # Extract days, hours, and minutes
    days = time_difference.days
    hours, remainder = divmod(time_difference.seconds, 3600)
    minutes, seconds = divmod(remainder, 60)
    
    return days, hours, minutes

def main():
    # Prompt the user for input
    date1 = input("Enter the first date (YYYY-MM-DD): ")
    date2 = input("Enter the second date (YYYY-MM-DD): ")
    
    # Calculate time difference
    days, hours, minutes = calculate_time_difference(date1, date2)
    
    # Display the result
    print(f"Time difference: {days} days, {hours} hours, {minutes} minutes")

if __name__ == "__main__":
    main()
```

```python
#Q9.

import pandas as pd

def main():
    # Prompt the user for file path, column name, and category order
    file_path = input("Enter the path to the CSV file: ")
    column_name = input("Enter the name of the column to convert: ")
    categories = input("Enter the categories in the desired order (comma-separated): ").split(",")

    # Read the CSV file
    try:
        df = pd.read_csv(file_path)
    except FileNotFoundError:
        print("File not found. Exiting...")
        return
    
# Convert the specified column to categorical
    if column_name in df.columns:
        df[column_name] = pd.Categorical(df[column_name], categories=categories, ordered=True)
        sorted_df = df.sort_values(by=column_name)
        
        # Display the sorted data
        print("Sorted Data:")
        print(sorted_df)
    else:
        print(f"Column '{column_name}' not found in the CSV file.")

if __name__ == "__main__":
    main()
```

```python
#Q10.

import pandas as pd
import matplotlib.pyplot as plt

def visualize_sales_data(file_path):
    # Load CSV data into a pandas DataFrame
    try:
        df = pd.read_csv(file_path)
    except FileNotFoundError:
        print("File not found. Please provide a valid file path.")
        return
    except pd.errors.EmptyDataError:
        print("The CSV file is empty.")
        return

    # Convert the 'Date' column to datetime
    df['Date'] = pd.to_datetime(df['Date'])

    # Group data by 'Date' and 'Product Category' and calculate the sum of sales
    sales_by_date_category = df.groupby(['Date', 'Product Category'])['Sales'].sum().unstack()
    
# Plot the stacked bar chart
    plt.figure(figsize=(10, 6))
    sales_by_date_category.plot(kind='bar', stacked=True)
    plt.title('Sales by Product Category Over Time')
    plt.xlabel('Date')
    plt.ylabel('Sales')
    plt.xticks(rotation=45)
    plt.legend(title='Product Category')
    plt.tight_layout()

    # Display the chart
    plt.show()

def main():
    file_path = input("Enter the path to the CSV file: ")
    visualize_sales_data(file_path)

if __name__ == "__main__":
    main()
```

```python
#Q11.

import pandas as pd
from scipy import stats

# Prompt user for file path
file_path = input("Enter the file path of the CSV file containing the student data: ")

# Read CSV file into a Pandas DataFrame
data = pd.read_csv(file_path)

# Calculate mean, median, and mode
mean = data['Test Score'].mean()
median = data['Test Score'].median()
mode = stats.mode(data['Test Score'])[0]

# Display results in a table
result_table = pd.DataFrame({
    'Statistic': ['Mean', 'Median', 'Mode'],
    'Value': [mean, median, ', '.join(map(str, mode))]
})

print(result_table)

```
